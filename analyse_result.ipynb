```python
import qsharp
import matplotlib.pyplot as plt
import numpy as np
qsharp.init(project_root = 'C:\\Users\\notip\\Epita\\Q#\\myQPE\\')

phi=(np.pi/10)

R = []
for i in range(1,150):
  r = qsharp.eval("PhaseEstimation.Run({}, {}, {})".format(i, phi, 1))
  phiEstim = -np.pi/2 - 2*np.arctan2(np.sqrt(r[1]),np.sqrt(r[0]))/i
  R.append(abs(phiEstim - phi))

plt.plot(R)
plt.show()

```

```python
R = []
for i in range(1,300):
  r = qsharp.eval("PhaseEstimation.Run({}, {}, {})".format(10, phi, i))
  phichap = -np.pi/(2*i)-2*np.arctan2(np.sqrt(r[1]),np.sqrt(r[0]))/10
  R.append(abs(phichap - phi))

plt.plot(R)
plt.show()
```
